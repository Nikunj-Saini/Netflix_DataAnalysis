### Now we will   make priorities in the vote_average section  in  below_average , average and popular, Not_popular

```python
def categorize_col(df, col, labels):

    # If already categorical / string, don't recalculate
    if not pd.api.types.is_numeric_dtype(df[col]):
        print(f"{col} is already categorized")
        return df

    edges = [
        df[col].min(),
        df[col].quantile(0.25),
        df[col].quantile(0.50),
        df[col].quantile(0.75),
        df[col].max()
    ]

    df[col] = pd.cut(
        df[col],
        bins=edges,
        labels=labels,
        duplicates='drop'
    )

    return df
```

```python
labels = ['not_popular', 'below_average', 'average', 'popular']

categorize_col(df, 'Vote_Average', labels)

df['Vote_Average'].unique()
```
